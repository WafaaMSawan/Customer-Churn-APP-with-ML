## Key Insights from EDA



* **Class Imbalance**: The class imbalance so evaluation metrics such as ROC-AUC are preferable over accuracy.
* **Contract Type**: Contract length is the strongest predictor of churn. Month-to-month customers exhibit very high churn rates compared to two-year contracts.
* **Customer Tenure**: Churn is concentrated among newer customers. The probability of churn decreases significantly as tenure increases.
* **Internet Service**: Customers with fiber optic internet show higher churn rates compared to DSL or customers without internet service.
* **Payment Method**: Electronic check users demonstrate elevated churn levels, potentially indicating less stable payment behavior.
* **Service Adoption** : Customers without additional services such as OnlineSecurity or TechSupport are significantly more likely to churn.

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
from sklearn.metrics import make_scorer, precision_score, recall_score, f1_score
param_grid_log = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga'],
    'max_iter': [100, 200, 500]
}
scoring_log = {
    'accuracy': 'accuracy',
    'precision_macro': make_scorer(precision_score, average='macro'),
    'recall_macro': make_scorer(recall_score, average='macro'),
    'f1_macro': make_scorer(f1_score, average='macro'),
    'roc_auc': 'roc_auc'
}
```

```python
from sklearn.model_selection import GridSearchCV

# Apply Grid Search with 5-fold CV
grid_search_log = GridSearchCV(logreg, param_grid_log, cv=5, scoring=scoring_log, refit='accuracy')
grid_search_log.fit(x_train, y_train)
```

```python
# Best parameters
print("Best parameters:", grid_search_log.best_params_)
```

```python
# Evaluate best model
best_model_log = grid_search_log.best_estimator_
y_pred_log = best_model_log.predict(x_test)
```

```python
# Get probability predictions
y_prob_log = best_model_log.predict_proba(x_test)[:,1]
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Compute ROC curve
fpr_log, tpr_log, thresholds = roc_curve(y_test, y_prob_log)
```

```python
# Compute AUC
auc_score_log = roc_auc_score(y_test, y_prob_log)
```

```python
# Plot
plt.figure(figsize=(6,5))
plt.plot(fpr_log, tpr_log, label=f"AUC = {auc_score_log:.3f}")
plt.plot([0,1], [0,1], linestyle='--')  # random classifier line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Logistic Regression")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import classification_report
print("\nClassification Report:\n", classification_report(y_test, y_pred_log))
```

```python
from sklearn.metrics import confusion_matrix
# Confusion Matrix
print("\nConfusion Matrix:\n")
cm = confusion_matrix(y_test, y_pred_log)
print(cm)
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=[0,1], yticklabels=[0,1])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

## XGBoost


```python
import xgboost as xgb
from sklearn.metrics import classification_report, roc_auc_score
```

```python
# Baseline XGBoost (No Imbalance Handling)

xgb_baseline = xgb.XGBClassifier(
    random_state=42,
    use_label_encoder=False,
    eval_metric="logloss"
)

xgb_baseline.fit(x_train, y_train)

y_pred_base = xgb_baseline.predict(x_test)
y_probs_base = xgb_baseline.predict_proba(x_test)[:, 1]

print(classification_report(y_test, y_pred_base))
print("ROC-AUC:", roc_auc_score(y_test, y_probs_base))
```

### XGBoost with handling class imbalance

```python
# Compute class counts

neg = np.sum(y_train == 0)
pos = np.sum(y_train == 1)

scale_pos_weight = neg / pos
print("scale_pos_weight =", scale_pos_weight)
```

### Using scale_pos_weight

```python
xgb_weighted = xgb.XGBClassifier(
    random_state=42,
    use_label_encoder=False,
    eval_metric="logloss",
    scale_pos_weight=scale_pos_weight
)

xgb_weighted.fit(x_train, y_train)

y_pred_w = xgb_weighted.predict(x_test)
y_prob_w = xgb_weighted.predict_proba(x_test)[:, 1]

print(classification_report(y_test, y_pred_w))
print("ROC-AUC:", roc_auc_score(y_test, y_prob_w))
```

## Feature Engineering for Naïve Bayes ,Random forest

```python
import numpy as np
import pandas as pd

def engineer_features(df):

    df_new = df.copy()

    # Derived from tenure and charges
    df_new['AvgMonthlySpend'] = df_new['TotalCharges'] / (df_new['tenure'] + 1)

    df_new['Tenure_squared'] = df_new['tenure'] ** 2

    df_new['IsNewCustomer'] = (df_new['tenure'] < 6).astype(int)

    # Service combinations and counts

    premium_cols = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport']
    df_new['PremiumServices'] = df_new[premium_cols].sum(axis=1)

    df_new['HasStreaming'] = (df_new['StreamingTV'] | df_new['StreamingMovies']).astype(int)

    has_internet = 1 - df_new['InternetService_No']
    df_new['InternetAndPremium'] = has_internet * df_new['PremiumServices']

    df_new['MonthlyPerService'] = df_new['MonthlyCharges'] / (df_new['TotalServices'] + 1)

    #Demographics & household

    df_new['FamilySize'] = df_new['Partner'] + df_new['Dependents']
    df_new['HasPartnerAndDependents'] = (df_new['Partner'] & df_new['Dependents']).astype(int)
    df_new['SeniorAlone'] = (df_new['SeniorCitizen'] & (df_new['Partner'] == 0)).astype(int)

    #Payment behaviour
    df_new['PaperlessAuto'] = (df_new['PaperlessBilling'] & df_new['PaymentMethod_Credit card (automatic)']).astype(int)

    #Contract & tenure interaction
    df_new['LongContractNew'] = ((df_new['Contract'] > 0) & (df_new['tenure'] < 12)).astype(int)

    return df_new
```

```python
#Apply the feature engineering
X_train_naive_rf = engineer_features(x_train)
X_test_naive_rf = engineer_features(x_test)
```
